```python
#pip install scipy
from scipy.stats import norm
#Usos comunes de SciPy:
#Estadísticas y pruebas: Como tests de hipótesis, pruebas t, análisis de varianza (ANOVA), etc.
#Optimización: Resolver problemas de optimización matemática.
#Cálculos numéricos: Integración numérica, ecuaciones diferenciales, interpolación.
#Álgebra lineal: Cálculo de determinantes, eigenvalores, factorización de matrices.
print(norm.cdf(1.23)) # resuelve P(z < 1.23), esto lo podemos hace con la tabla de la Distribución Acumulativa Normal
print(norm.cdf(0) - norm.cdf(-0.67)) # resuelve P(−0.67 < z < 0) o print(norm.cdf(0) - (1 - norm.cdf(0.67)))
print((1 - norm.cdf(2.30)))#Es decir, la probabilidad de que una variable aleatoria normal estándar sea mayor que 2.30.
print(norm.cdf(-0.33) - norm.cdf(1.45))
print((1 - norm.cdf(-2.10)))# resuelve P(z >-2.10)
print(norm.cdf(-3.5))  
```

```python
from scipy.stats import uniform
#parametrizamos 
a=0
b=10
#P(X < 3)
print(uniform.cdf(3,a,b-a)) #o (3-a)/(b-a)
#P(X > 6)
print(1-uniform.cdf(6,a,b-a)) #o 1-(6-a)/(b-a) 
#P(3 < X < 8)
print(uniform.cdf(8,a,b-a)-uniform.cdf(3,a,b-a)) #o (8-a)/(b-a)- (3-a)/(b-a)
```

```python
from sympy import * #importa todas la librerias de sympy
x,y = symbols('x y')
fxy = (2/3) * (x + y) * exp(-x)
Ex= integrate(x*integrate(fxy,(y,0,1)),(x,0,1))
Ey= integrate(y*integrate(fxy,(x,0,1)),(y,0,1))
Ex2= integrate(x**2*integrate(fxy,(y,0,1)),(x,0,1))
Ey2= integrate(y**2*integrate(fxy,(x,0,1)),(y,0,1))
Exy= integrate(x*y*integrate(fxy,(y,0,1)),(x,0,1))
cov=Exy-Ex*Ey
varx=Ex2-Ex**2
vary=Ey2-Ey**2
rho=cov/(sqrt(varx*vary))
rho.evalf()
```
