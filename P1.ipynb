```python
import numpy as np              
import pandas as pd 
import os #ya viene en python
#py -m pip install pandas numpy wget zipfile sympy 
os.listdir()# Obtiene la lista de archivos y directorios en el directorio actual
Data=pd.read_stata('B1.dta')#read_stata(): Es una funcin de pandas que permite leer archivos de Stata (.dta) y convertirlos en un DataFrame.
# Mostrar las primeras 5 filas del DataFrame
print(Data)
#print(Data.head())  # Tambin puedes usar Data.head() sin print
```

```python
X=Data[['x3','x4','x5','x6','x7','x8']]
X # mostramos valores del array X
```

```python
X.shape # vemos las filas y columnas de X
```

```python
F=X.shape[0] # obtener el nmero de filas del objeto.
F
Unos = np.ones((F, 1)) #Crea una matriz de unos de tamao (F, 1), donde F es el nmero de filas. Esto te da una columna de unos con F filas.
Unos
#Mo:matriz indempotente til
Mo = np.eye(F) - 1/F * Unos @ Unos.T
#np.eye(F): Crea una matriz identidad de tamao F x F.
#Unos @ Unos.T: Crea una matriz F x F con todos sus elementos iguales a 1.
#1/F * Unos @ Unos.T: Divide todos los elementos de la matriz de unos por F.
#La operacin final np.eye(F) - 1/F * Unos @ Unos.T resta la matriz de unos escalada de la matriz identidad.
#En Python, el operador @ se utiliza para realizar una multiplicacin de matrice
Mo#tiene dimension 21x21
```

```python
#Matriz de Varianza y covarianzas 
#1 / (F - 1) hacemos esto para convertirlo a la varianza muestral, a travez de las muestras finitas.
S = X.T @ Mo @ X * 1 / (F - 1)
S
#Para corrobarar utilizamos X.cov(), el cual tiene q ser igual a la matriz S
```

```python
#Ahora resolvemos la segunda proposicin 
np.diagonal(S).sum() #extraemos elementos de la diagonal y los sumamos #Tambien podemos utilizas np.trace(S), obtenemos la traza
print(np.linalg.det(S))
#np.linalg es el submdulo de NumPy que se utiliza para operaciones de lgebra lineal.
#det(S) calcula el determinante de la matriz S. El determinante de una matriz es un valor escalar que puede ser usado para analizar la matriz, especialmente para determinar si es invertible (si el determinante es distinto de cero, la matriz es invertible).
Auto = np.linalg.eig(S)#autovalores de S
#np.linalg.eig() es una funcin de NumPy que se utiliza para calcular los autovalores y autovectores de una matriz cuadrada S.
#Auto = np.linalg.eig(S):Nos da los autovalores y los autovectores
print(Auto[0].sum(),Auto[0].prod())
#Auto[0]nos da solo los autovalores
#Nos da el resultado de la suma y los productos de los autovalores
```

```python
D = np.diag(np.diag(S)**(-0.5))
D
#np.diag(S): Extrae los elementos de la diagonal principal de la matriz S, devolviendo un vector con esos valores.
#np.diag(S)**(-0.5): Eleva cada uno de los elementos de la diagonal (que est ahora en forma de un vector) a la potencia de 0.5, es decir, calcula la raz cuadrada inversa de cada elemento de la diagonal. Esto es equivalente a tomar el inverso de la raz cuadrada de cada valor en la diagonal.
#np.diag(np.diag(S)**(-0.5)): Finalmente, vuelve a crear una matriz diagonal con estos valores elevados, es decir, los valores en la diagonal de la nueva matriz sern los inversos de las races cuadradas de los valores en la diagonal de S, y todos los elementos fuera de la diagonal sern cero.

R=D@S@D
R
#El 0, 1, 2, 3, 4, 5 que ves tanto arriba como al lado izquierdo son solo los ndices que indican la posicin de los elementos dentro de la matriz. Esto te ayuda a ubicar cada valor dentro de la matriz cuando trabajas con matrices grandes o cuando necesitas acceder a elementos especficos de ellas.
```

```python
import sympy  as sp
#SimPy es un paquete que permite la simulacin de sistemas de eventos discretos. Algunos de los conceptos clave que maneja incluyen:
    #Procesos: Entidades que representan tareas que ocurren a lo largo del tiempo.
    #Recursos: Cosas que pueden ser compartidas entre varios procesos, como mquinas, personal o cualquier otra capacidad limitada.
    #Eventos: Momentos especficos en el tiempo en los que algo sucede en el sistema.
alpha = sp.symbols('alpha')
#sp.symbols(): Es una funcin de SymPy que permite declarar variables simblicas. Es como declarar una variable matemtica en una expresin, en lugar de usar un valor numrico.
#'alpha': Es el nombre de la variable simblica que ests creando. En este caso, se est creando una variable simblica llamada alpha.
x1,x2,x3 = sp.symbols('x1 x2 x3') #definimos 3 simbolos ms x1,x2,x3
A= sp.Matrix([[3,0,2],[-3,-3,-3],[alpha,0,4]])#ultizamos sp.Matrix, pues la matriz contiene sumbolos
b= sp.Matrix([[4],[0],[-4]])
X= sp.Matrix([[x1],[x2],[x3]])
Am=A-6*sp.eye(3)#Am es la matriz modificada #eye(3)matriz identidad de 3x3
Amd=Am.det()#obtenemos la determinate de la matriz modificada
res=sp.solve(Amd,alpha)#solve se utiliza para resolver ecuaciones simblicas o sistemas de ecuaciones.
res
alpha = res[0]# Asignar la primera solucin a alpha osea 3
A= sp.Matrix([[3,0,2],[-3,-3,-3],[alpha,0,4]])
A
rf=sp.solve((A-6*sp.eye(3))@X-b,['x1', 'x2','x3'])# resolvemos el sistema de ecuaciones 
rf
# los resultados dependeran de x3 
```

```python
import numpy as np
A= np.array([[1,1,2],[0,-1,-4],[1,1,1]])
B=np.array([[1,0,1],[0,-1,0]])
C=np.array([[0,2,1],[1,1,1]])
BT=np.transpose(B)
resol=A-BT@C
det_resol=np.linalg.det(resol)
det_resol # si la determinanate es dfierente de 0 es inversible o no singular

# Funcin para calcular el determinante de una submatriz
def determinante_submatriz(matriz, fila, columna):
    submatriz = np.delete(matriz, fila, axis=0)  # Eliminar fila
    submatriz = np.delete(submatriz, columna, axis=1)  # Eliminar columna
    return np.linalg.det(submatriz)

# Funcin para calcular la matriz de cofactores
def matriz_cofactores(matriz):
    filas, columnas = matriz.shape
    cofactores = np.zeros((filas, columnas))
    
    for i in range(filas):
        for j in range(columnas):
            # Cofactor de cada elemento
            cofactores[i, j] = (-1) ** (i + j) * determinante_submatriz(matriz, i, j)
    
    return cofactores
    
# Calculamos la matriz de cofactores
cofactores_resol = matriz_cofactores(resol)
cofactores_resol # matriz de cofactores
cofactores_resol_T=np.transpose(cofactores_resol)
cofactores_resol_T
resol_final=(cofactores_resol_T/(-1))
resol_final
resol_directa = np.linalg.inv(resol) if np.linalg.det(A) != 0 else "La matriz no tiene inversa" # en caso la matriz sea difretne de 0 (!=0) tendra inversa.
resol_directa
```

```python
import sympy as sp
b,a,d,c,alpha=sp.symbols('b a d c alpha')
A=sp.Matrix([[7*b,-a,0,0],[7*d,-c,0,0],[0,0,3,1],[0,0,0,alpha+2]])
B=sp.Matrix([[a,b],[c,d]])
cond1=sp.Eq(A.det()-1,0)#sp.Eq() toma dos argumentos y establece que la expresin de la izquierda es igual a la expresin de la derecha.
cond2=sp.Eq(B.det(),5)#A.det() calcula el determinante de la matriz 
res=sp.solve([cond1,cond2],(b,a,d,c,alpha))
res
```
